```python
#Ejercicio 1
class Empleado:
    """clase base para representar un empleado general."""
    def __init__(self, nombre, empresa, salario_base):
        self.nombre = nombre
        self.empresa = empresa
        self.salario_base = salario_base

    def calcular_salario(self):
        return self.salario_base

    def descripcion(self):
        return f"{self.nombre} trabaja en {self.empresa}"

class GerenteVentas(Empleado):
    """subclase que añade un bono por ventas."""
    def __init__(self, nombre, empresa, salario_base, bono_ventas):
        super().__init__(nombre, empresa, salario_base)
        self.bono_ventas = bono_ventas

    def calcular_salario(self):
        return self.salario_base + self.bono_ventas

class ObreroPorHora(Empleado):
    """subclase que calcula salario basado en horas trabajadas."""
    def __init__(self, nombre, empresa, horas_trabajadas, valor_hora):
        super().__init__(nombre, empresa,0) 
        self.horas_trabajadas = horas_trabajadas
        self.valor_hora = valor_hora

    def calcular_salario(self):
        return self.horas_trabajadas * self.valor_hora

gerente = GerenteVentas("Ana Pérez", "UA", 2000000, 300000)
obrero = ObreroPorHora("Pedro Soto", "UA", 80, 8000)
print("Gerente Ventas:", gerente.nombre, "- Salario:", gerente.calcular_salario()) 
print("Obrero PorHora:", obrero.nombre, "- Salario:", obrero.calcular_salario()) 
```

```python
#Ejercicio 2
class Vehiculo:
    """Clase base que define atributos comunes de cualquier vehículo."""
    def __init__(self, marca, modelo, año):
        self.marca = marca
        self.modelo = modelo
        self.año = año

class Electrico:
    """Mixin que añade funcionalidad de motor eléctrico."""
    def __init__(self, bateria_kwh):
        self.bateria_kwh = bateria_kwh

class Combustion:
    def __init__(self, cilindrada_cc):
        self.cilindrada_cc = cilindrada_cc

class Auto(Vehiculo):
    def ficha_tecnica(self):
        print(f"Auto: {self.marca} {self.modelo} ({self.año})")

class Moto(Vehiculo):
    def ficha_tecnica(self):
        print(f"Moto: {self.marca} {self.modelo} ({self.año})")

class AutoElectrico(Auto, Electrico):
    def __init__(self, marca, modelo, año, bateria_kwh):
        Vehiculo.__init__(self, marca, modelo, año)
        Electrico.__init__(self, bateria_kwh)

    def ficha_tecnica(self):
        print(f"FICHA ELÉCTRICA: {self.marca} {self.modelo}")
        print(f"- Año: {self.año}")
        print(f"- Capacidad Batería: {self.bateria_kwh} kWh")

mi_tesla = AutoElectrico("Tesla", "Model 4", 2025, 80)
mi_tesla.ficha_tecnica()
print("Orden de Resolución de Métodos (MRO):")
for clase in AutoElectrico.__mro__:
    print(clase)
```

```python
#ejercicio3
from abc import ABC, abstractmethod

class MetodoPago(ABC):
    """Interfaz abstracta para metodos de pago."""
    @abstractmethod
    def procesar(self, monto):
        """Metodo para procesar el pago."""
        pass

    @abstractmethod
    def reembolsar(self, monto):
        pass

class PagoTarjeta(MetodoPago):
    """Implementacion para pagos con tarjeta."""
    def __init__(self, numero_tarjeta, titular):
        self.numero_tarjeta = numero_tarjeta
        self.titular = titular

    def procesar(self, monto):
        print(f"Pagando {monto} con tarjeta {self.numero_tarjeta}")

    def reembolsar(self, monto):
        print(f"Reembolsando {monto} a {self.titular}")

class PagoTransferencia(MetodoPago):
    """Implementacion para transferencias bancarias."""
    def __init__(self, cuenta_origen, cuenta_destino):
        self.cuenta_origen = cuenta_origen
        self.cuenta_destino = cuenta_destino

    def procesar(self, monto):
        print(f"Transferencia de {monto} desde {self.cuenta_origen}")

    def reembolsar(self, monto):
        print(f"Devolviendo {monto} a {self.cuenta_origen}")

class PagoEfectivo(MetodoPago):
    """Implementacion para pagos en efectivo."""
    def procesar(self, monto):
        print(f"Pago de {monto} recibido en efectivo")

    def reembolsar(self, monto):
        print(f"Devolviendo {monto} en efectivo")

def ejecutar_pago(metodo, monto):

    metodo.procesar(monto)

pago_test = PagoTarjeta("4540-0000", "Ana Perez")
ejecutar_pago(pago_test, 25000)
```

```python
#ejercicio4
from abc import ABC, abstractmethod

class MetodoPago(ABC):
    """Interfaz abstracta para definir el contrato de pagos."""
    
    @abstractmethod
    def procesar(self, monto):
        """Metodo obligatorio para realizar un cobro."""
        pass

    @abstractmethod
    def reembolsar(self, monto):
        """Metodo obligatorio para devolver dinero."""
        pass

class PagoTarjeta(MetodoPago):
    """Implementacion de pago con tarjeta de credito/debito."""
    def __init__(self, numero_tarjeta, titular):
        self.numero_tarjeta = numero_tarjeta
        self.titular = titular

    def procesar(self, monto):
        print(f"Procesando pago de ${monto} con tarjeta {self.numero_tarjeta}")

    def reembolsar(self, monto):
        print(f"Reembolsando ${monto} a la cuenta de {self.titular}")

class PagoTransferencia(MetodoPago):
    """Implementacion de pago via transferencia bancaria."""
    def __init__(self, cuenta_origen, cuenta_destino):
        self.cuenta_origen = cuenta_origen
        self.cuenta_destino = cuenta_destino

    def procesar(self, monto):
        print(f"Transferencia de ${monto} desde la cuenta {self.cuenta_origen}")

    def reembolsar(self, monto):
        print(f"Devolviendo ${monto} a la cuenta {self.cuenta_origen}")

class PagoEfectivo(MetodoPago):
    """Implementacion de pago en efectivo."""
    def procesar(self, monto):
        print(f"Pago de ${monto} registrado en efectivo por caja")

    def reembolsar(self, monto):
        print(f"Devolucion de ${monto} realizada en efectivo")

def ejecutar_pago(metodo, monto):
    metodo.procesar(monto)

# Creamos diferentes objetos de pago
tarjeta = PagoTarjeta("4540-1234", "Ana Perez")
efectivo = PagoEfectivo()

ejecutar_pago(tarjeta, 50000)
ejecutar_pago(efectivo, 15000)
```
