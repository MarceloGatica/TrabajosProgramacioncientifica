```python
#EJERCICIO1
import pandas as pd
df = pd.read_csv('clima_talca_enero2024.csv')
print("Primeras 5 filas")
display(df.head(5))
print("Últimas 3 filas")
display(df.tail(3))
print("Información del DataFrame")
df.info()
print("Estadísticas descriptivas")
display(df.describe())
print("Filas y columnas")
display(df.shape)
```

```python
#EJERCICIO2
nulos_por_columna = df.isnull().sum()
print("\n conteo de nulos por columna")
print(nulos_por_columna)

porcentaje_nulos = (nulos_por_columna / len(df)) * 100
print("\n porcentaje de nulos por columna")
print(porcentaje_nulos)

columnas_mas_15 = porcentaje_nulos[porcentaje_nulos > 15]
print("\n columnas con más del 15% de nulos")
print(columnas_mas_15)
```

```python
#EJERCICIO3
df_limpio = df.copy()

df_limpio['temp_max'] = df_limpio['temp_max'].fillna(df_limpio['temp_max'].mean())
df_limpio['temp_min'] = df_limpio['temp_min'].fillna(df_limpio['temp_min'].mean())

df_limpio['humedad_pct'] = df_limpio['humedad_pct'].fillna(df_limpio['humedad_pct'].median())

df_limpio['precipitacion_mm'] = df_limpio['precipitacion_mm'].fillna(0.0)

df_limpio = df_limpio.dropna(subset=['condicion'])

print("nulos despues de la limpieza")
print(df_limpio.isnull().sum())
```

```python
#EJERCICIO4
display(df_limpio[['fecha', 'temp_max', 'condicion']])

display(df_limpio.iloc[:5])

display(df_limpio.loc[14])

display(df_limpio[df_limpio['temp_max'] >= 34])

display(df_limpio[df_limpio['condicion'] == 'Lluvioso'])

display(df_limpio[(df_limpio['temp_max'] > 30) & (df_limpio['humedad_pct'] > 50)])

df_limpio['amplitud_termica'] = df_limpio['temp_max'] - df_limpio['temp_min']
display(df_limpio.head())
```

```python
#EJERCICIO5
# 1. Día más caluroso del mes
dia_mas_caluroso = df_limpio.loc[df_limpio['temp_max'].idxmax()]
print("--- Día más caluroso ---")
print(dia_mas_caluroso[['fecha', 'temp_max']])

# 2. Día con mayor precipitación
dia_mayor_lluvia = df_limpio.loc[df_limpio['precipitacion_mm'].idxmax()]
print("\n--- Día con mayor precipitación ---")
print(dia_mayor_lluvia[['fecha', 'precipitacion_mm']])

# 3. Conteo de días para cada condición climática
print("\n--- Frecuencia de condiciones climáticas ---")
print(df_limpio['condicion'].value_counts())

# 4. Temperatura máxima y mínima promedio del mes
print("\n--- Temperaturas promedio del mes ---")
print(f"Máxima promedio: {df_limpio['temp_max'].mean():.2f}")
print(f"Mínima promedio: {df_limpio['temp_min'].mean():.2f}")

# 5. Días duplicados en el dataset
duplicados = df_limpio.duplicated().sum()
print(f"\n--- Días duplicados encontrados: {duplicados} ---")

# 6. Resumen final con describe() solo de las columnas numéricas
print("\n--- Resumen descriptivo final ---")
display(df_limpio.describe())
```
