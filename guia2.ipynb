```python
from functools import reduce
import random
```

```python
# Actividad 1
ecuacion = lambda x: x**2 + 2*x + 1
valores = [2, 4, 6]
resultados = [ecuacion(v) for v in valores]
print("resultados:", resultados)
# Conclusion:lambda calculo los valores exactos del polinomio (9, 25 y 49).
```

```python
#Actividad 2 
velocidades = [10, 20, 30, 40, 50]
velocidades_ms = list(map(lambda x: x / 3.6, velocidades))
print("velocidades en m/s:", velocidades_ms)
# COnclusion map aplico la fórmula a toda la lista, convirtiendo las velocidades a m/s.
```

```python
#Actividad 3 – Filtrado con filter() 
valores = [2, 5, 8, 11, 14, 17, 20]
# Un numero es par si el resto de su división por 2 es 0 (x % 2 == 0)
pares = list(filter(lambda x: x % 2 == 0, valores))

print("numeros pares:", pares)
# Conclusion: filter depuro la lista original aislando únicamente los numeros pares.
```

```python
#Actividad 4 – Agregación con reduce()
valores = [2, 3, 4, 5]
producto_total = reduce(lambda x,y: x * y, valores)

print("producto total:", producto_total)
# Conclusion reduce multiplico la lista hasta obtener el producto total (120).
```

```python
#Actividad 5 – Pipeline de Procesamiento
data = [12, 15, 22, 35, 41, 8, 3, 55, 62, 90]

# Implementación del pipeline:
#  filter: Mantiene solo los valores mayores o iguales a 10 
#  map: Toma esos valores filtrados y los multiplica por 2 
#  reduce: suma todos los resultados finales 

resultado = reduce(lambda x, y: x + y, 
                   map(lambda x: x * 2, 
                       filter(lambda x: x >= 10, data))) 

print("resultado :", resultado)
# conclusion el pipeline filtro, multiplico por 2 y sumo los datos en un solo paso.
```

```python
# Actividad 6
data = [random.randint(0, 100) for _ in range(1000000)]

# filtramos los  valores mayores a 50
filtrados = filter(lambda x: x > 50, data)

# elevamos los valores al cuadrado (usamos list() para saber el largo)
cuadrados = list(map(lambda x: x**2, filtrados))

# calculamos el promedio final sumando todo y dividiendo por la cantidad de elementos
suma_total = reduce(lambda x, y: x + y, cuadrados)
promedio_final = suma_total / len(cuadrados)

print("promedio final :", promedio_final)
# Conclusion: se procesaron, transformaron y promediaron miles de registros simulados.
```

```python
# Simulamos los sensores
sensores = [random.uniform(10, 80) for _ in range(500000)]

# eliminar valores fuera del rango 10-70°C
sensores2 = filter(lambda x: 10 <= x <= 70, sensores)

# convertir a Kelvin (sumando 273.15)
kelvin = list(map(lambda x: x + 273.15, sensores2))

# calcular temperatura promedio
suma_kelvin = reduce(lambda x, y: x + y, kelvin)
promedio_kelvin = suma_kelvin / len(kelvin)

print("temperatura promedio en kelvin :", promedio_kelvin)
# Conclusion: se procesaron, transformaron y promediaron miles de registros simulados.
```
