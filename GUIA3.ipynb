```python
#Ejercicio 1
from functools import total_ordering

# Usa total_ordering para derivar los demas operadores de comparacion
@total_ordering
class Producto:
    def __init__(self, nombre, precio, stock):
        self.nombre = nombre
        self.precio = precio
        self.stock = stock

    def __str__(self):
        return f"{self.nombre} ${self.precio} ({self.stock} unidades)"

    def __repr__(self):
        return f"Producto('{self.nombre}', {self.precio}, {self.stock})"

    # Comparaciones basadas en el precio
    def __eq__(self, otro):
        return self.precio == otro.precio

    def __lt__(self, otro):
        return self.precio < otro.precio

    # Aritmetica
    def __add__(self, otro):
        if self.nombre == otro.nombre:
            nuevo_stock = self.stock + otro.stock
            return Producto(self.nombre, self.precio, nuevo_stock)
        else:
            raise ValueError("Los productos deben tener el mismo nombre")

    def __mul__(self, n):
        nuevo_precio = self.precio * n
        return Producto(self.nombre, nuevo_precio, self.stock)


# Pruebas
if __name__ == "__main__":
    p1 = Producto('Arroz', 1200, 50)
    p2 = Producto('Arroz', 900, 30)
    p3 = Producto('Leche', 1200, 20)

    print(p1)
    print(repr(p2))
    print(p1 == p3)
    print(p2 < p1)
    print(sorted([p1, p2, p3])) 
    
    combinado = p1 + p2
    print(combinado)
    
    oferta = p1 * 0.8
    print(oferta)
```

```python
#Ejercicio 2 
class CuentaBancaria:
    def __init__(self, titular, saldo):
        self.titular = titular
        self.saldo = saldo

    # Getter y Setter para titular
    @property
    def titular(self):
        return self._titular

    @titular.setter
    def titular(self, valor):
        if len(valor) < 3:
            raise ValueError("El titular debe tener al menos 3 caracteres")
        self._titular = valor

    # Getter y Setter para saldo
    @property
    def saldo(self):
        return self._saldo

    @saldo.setter
    def saldo(self, valor):
        if valor < 0:
            raise ValueError("El saldo no puede ser negativo")
        self._saldo = valor

    # Operaciones (usan el setter implícitamente al sumar/restar a self.saldo)
    def depositar(self, monto):
        if monto > 0:
            self.saldo += monto

    def retirar(self, monto):
        if self.saldo - monto < 0:
            raise ValueError("Saldo insuficiente")
        self.saldo -= monto

    # Representaciones
    def __str__(self):
        return f"Cuenta de {self.titular} | Saldo: ${int(self.saldo)}"

    def __repr__(self):
        return f"CuentaBancaria(titular='{self.titular}', saldo={int(self.saldo)})"

    # Propiedad calculada
    @property
    def aprobada(self):
        return self.saldo >= 10000


# Pruebas
if __name__ == "__main__":
    cuenta = CuentaBancaria('Ana López', 50000)
    print(cuenta)
    
    cuenta.depositar(20000)
    print(cuenta.saldo)
    
    cuenta.retirar(15000)
    print(cuenta)
    
    try:
        cuenta.saldo = -1000
    except ValueError as e:
        print(e)
        
    try:
        CuentaBancaria('AB', 0)
    except ValueError as e:
        print(e)
        
    print(cuenta.aprobada)
```

```python
#Ejercicio 3
import math

class Figura:
    def __init__(self, color):
        self.color = color

    # Métodos abstractos
    def area(self):
        raise NotImplementedError("Método abstracto")

    def perimetro(self):
        raise NotImplementedError("Método abstracto")

    def __str__(self):
        # Formateo a 2 decimales según la salida esperada en la guía
        return f"Figura {self.color} | Área: {self.area():.2f} | Perímetro: {self.perimetro():.2f}"

    def __repr__(self):
        return f"Figura('{self.color}')"

    # Bonus: permite ordenar las figuras por área usando sorted()
    def __lt__(self, otra):
        return self.area() < otra.area()


class Circulo(Figura):
    def __init__(self, color, radio):
        super().__init__(color)
        if radio <= 0:
            raise ValueError("El radio debe ser mayor a 0")
        self.radio = radio

    def area(self):
        return math.pi * (self.radio ** 2)

    def perimetro(self):
        return 2 * math.pi * self.radio

    def __repr__(self):
        return f"Circulo('{self.color}', {self.radio})"


class Rectangulo(Figura):
    def __init__(self, color, ancho, alto):
        super().__init__(color)
        if ancho <= 0 or alto <= 0:
            raise ValueError("Ancho y alto deben ser mayores a 0")
        self.ancho = ancho
        self.alto = alto

    def area(self):
        return self.ancho * self.alto

    def perimetro(self):
        return 2 * (self.ancho + self.alto)

    @property
    def es_cuadrado(self):
        return self.ancho == self.alto

    def __repr__(self):
        return f"Rectangulo('{self.color}', {self.ancho}, {self.alto})"


# Pruebas
if __name__ == "__main__":
    c = Circulo('rojo', 5)
    r = Rectangulo('azul', 4, 6)
    cuad = Rectangulo('verde', 5, 5)

    print(c)
    print(r)

    print(cuad.es_cuadrado)
    print(r.es_cuadrado)

    # Polimorfismo
    figuras = [c, r, cuad]
    for f in figuras:
        print(f"Área: {f.area():.2f}")

    # Prueba del Bonus de ordenamiento
    print("\nOrdenadas por área:")
    for f in sorted(figuras):
        print(f"{f.__class__.__name__}: {f.area():.2f}")
```

```python
#Ejercicio 4
class Inventario:
    def __init__(self):
        self._productos = []

    # --- Métodos de secuencia ---
    def agregar(self, producto):
        self._productos.append(producto)

    def __len__(self):
        return len(self._productos)

    def __getitem__(self, i):
        return self._productos[i]

    def __contains__(self, producto):
        return producto in self._productos

    def __iter__(self):
        return iter(self._productos)

    # --- Representación y análisis ---
    def __str__(self):
        return "\n".join(str(p) for p in self._productos)

    def __repr__(self):
        return f"Inventario(n_productos={len(self._productos)})"

    @property
    def total_valor(self):
        return sum(p.precio * p.stock for p in self._productos)

    @property
    def mas_caro(self):
        if not self._productos:
            return None
        # max() funciona directo porque Producto tiene el método __lt__
        return max(self._productos)

    def buscar(self, nombre):
        return [p for p in self._productos if nombre.lower() in p.nombre.lower()]


# Pruebas
if __name__ == "__main__":
    # Asegúrate de tener la clase Producto del Ejercicio 1 en el mismo script
    
    inv = Inventario()
    inv.agregar(Producto('Arroz', 1200, 50))
    inv.agregar(Producto('Leche', 950, 30))
    inv.agregar(Producto('Pan', 1800, 100))
    inv.agregar(Producto('Paneton', 3500, 10))

    print(len(inv))
    print(inv[0])

    p_buscar = Producto('Leche', 950, 30)
    print(p_buscar in inv)

    print("\nIterando:")
    for p in inv:
        print(p)

    print(f"\nValor total: ${inv.total_valor}")
    print(f"Más caro: {inv.mas_caro}")

    print("\nResultados de búsqueda ('pan'):")
    resultados = inv.buscar('pan')
    for r in resultados:
        print(r)
```
